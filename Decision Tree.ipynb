# Decision Tree

Un Decision Tree, o albero decisionale, è un modello di apprendimento automatico utilizzato per problemi di classificazione e regressione. Si tratta di una struttura ad albero in cui ogni nodo rappresenta una decisione basata su una caratteristica (o attributo), e ogni ramo rappresenta un possibile risultato di quella decisione. I nodi foglia dell'albero contengono le etichette di classe o i valori di output.

## Ecco come funziona un Decision Tree:

- **Radice dell'Albero**: La caratteristica (attributo) che viene scelta come nodo radice è quella che meglio suddivide il set di dati iniziale in sottoinsiemi più omogenei in termini di classe (per la classificazione) o di valore target (per la regressione).

- **Nodi Interni**: Ciascun nodo interno dell'albero rappresenta una decisione basata su un certo attributo. L'obiettivo è suddividere il set di dati in modo che i sottoinsiemi risultanti siano più puri, cioè contengono istanze appartenenti a una sola classe (per la classificazione) o hanno valori target simili (per la regressione).

- **Rami**: I rami rappresentano le diverse opzioni o valori che l'attributo può assumere. Ad ogni nodo, i dati vengono suddivisi in base al valore dell'attributo.

- **Nodi Foglia**: I nodi foglia contengono le etichette di classe (per la classificazione) o i valori di output (per la regressione). Sono i risultati finali della sequenza di decisioni prese dai nodi interni.

- **Criterio di Suddivisione**: Il criterio utilizzato per decidere quale attributo suddivide il dataset è essenziale. Nel tuo codice, criterion='entropy' indica che l'entropia viene utilizzata come misura della purezza del subset.

- **Pruning (Poda)**: Alcuni alberi decisionali possono diventare molto complessi e suscettibili all'overfitting. La poda (pruning) è una tecnica che consiste nel rimuovere rami o nodi non necessari per migliorare le prestazioni del modello.

### Vantaggi: 
Sono facili da interpretare e visualizzare. Non richiedono molte preparazioni dei dati, come la normalizzazione delle feature.

### Svantaggi: 
Possono essere suscettibili all'overfitting, specialmente su dataset piccoli. La creazione di alberi complessi può portare a modelli difficili da interpretare.

Il Decision Tree è un modello potente e versatile, spesso utilizzato per la sua interpretabilità e facilità di implementazione. Inoltre, può essere utilizzato sia per problemi di classificazione che di regressione.

```python
# Importiamo librerie
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree
```

```python
# Prendiamo in input i dati
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values
```

ll codice implementa un modello di classificazione basato su un Decision Tree che cerca di prevedere se un utente effettuerà o meno un acquisto in base alle caratteristiche dell'età e del salario stimato. La matrice di confusione e altre metriche di valutazione forniscono informazioni sulle prestazioni del modello.

```python
# Il dataset è diviso in un set di addestramento e un set di test utilizzando la funzione train_test_split di scikit-learn.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Le caratteristiche del set di addestramento vengono standardizzate utilizzando la StandardScaler di scikit-learn.
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Viene creato un classificatore Decision Tree utilizzando l'entropia come criterio di suddivisione e il seed (random_state) per la riproducibilità dei risultati.
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

```python
# Il modello Decision Tree è utilizzato per prevedere l'output del set di test.
y_pred = classifier.predict(X_test)
```

```python
# Viene utilizzata la matrice di confusione per valutare le prestazioni del modello Decision Tree.
cm = confusion_matrix(y_test, y_pred)
```

```python
# Visualizzazione dell'albero
plt.figure(figsize=(100, 80))
plot_tree(classifier, feature_names=['Age', 'EstimatedSalary'], class_names=['Not Purchased', 'Purchased'], filled=True)
plt.show()
```
